```python
from bs4 import BeautifulSoup

# Tên file HTML của bạn
file_path = 'construct.html'

try:
    # Mở và đọc nội dung file
    with open(file_path, 'r', encoding='utf-8') as f:
        html_content = f.read()

    # Tạo đối tượng BeautifulSoup để phân tích HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Tìm tất cả các thẻ <a> có thuộc tính href
    all_links = soup.find_all('a', href=True)

    # In ra giá trị href của mỗi thẻ tìm được
    print(f"Đã tìm thấy {len(all_links)} đường link:")
    for link in all_links:
        print(link['href'])

except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file tại '{file_path}'")
```
